```python
import networkx as nx
from matplotlib import pyplot as plt

G = nx.Graph()

G.add_node('Yutong')
G.add_nodes_from(['Cassidy', 'Comfort'])
nx.set_node_attributes(G, {'Cassidy': {'undergrad_US': True}, 'Yutong': {'undergrad_US': False}, 'Comfort': {'undergrad_US': True}})


G.add_edge('Yutong', 'Cassidy')
G.add_edges_from([('Cassidy', 'Comfort'), ('Yutong', 'Comfort')])

node_colors = ['green' if G.nodes[n]['undergrad_US'] else 'red' for n in G.nodes]

pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_color=node_colors, node_size=700, font_size=10)
```
